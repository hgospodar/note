# plt绘图组成部分分析
一个典型的plt绘图如下所示
<div align=center><img src="https://matplotlib.org/stable/_images/anatomy.png" style="zoom:40%" /></div>

所有的绘图都需要在**figure**上进行，使用`plt.figure`创建

**figure**里面是多个子图，称为**Axes**，使用`fig.subplots`创建

一个**Axes**由两个坐标轴组成，分别为**ax.xaxis**&**ax.yaxis**

图形由**Axes**的`ax.plot`&`ax.scatter`等函数创建，不同的函数名表示不同的图标类型。此函数接受横纵坐标集合，将图形绘制在**Axes**中.

# plt 用法一
可以直接将`plt`当作一个`axis`使用，如下代码所示

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5,6])
y = np.array([2,3,4,5,6,7])
plt.scatter(x,y)
plt.show()
````

# plt 用法2
根据一般的plt模块使用方式，应该遵循以下步骤
```mermaid
graph TB
    a(导入plt模块)
    b[新建figure]
    c[在figure中新建axies]
    d[在axies中导入横纵坐标值]
    e(展示figure)
    a-->b-->c-->d-->e
```
如下面代码所示

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5,6])
y = np.array([2,3,4,5,6,7])

# 下面开始绘图
# subplots返回一个新画布fig，和一个缺省的新绘图视窗ax
fig, ax = plt.subplots()
ax.scatter(x,y)
````
